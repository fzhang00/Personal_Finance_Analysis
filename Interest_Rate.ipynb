```python
idx_smp = [-1, -3, -20, -60, -120, -260]
yield_t_smp = (yield_t_df.iloc[idx_smp]).T
```

```python
# prompt: plot yield_t_smp, one column per line. use line with marker on data points. Use rainbow color sequence

import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np

# Assuming yield_t_smp is defined as in the previous example
# Create the plot with a rainbow color sequence
colors = plt.cm.plasma(np.linspace(0, 1, len(yield_t_smp)))

fig, ax = plt.subplots(figsize=(12, 6))

for i, col in enumerate(yield_t_smp.columns):
    ax.plot(yield_t_smp.index, yield_t_smp[col], marker='o', linestyle='-', color=colors[i], label=col)

# Customize the plot
ax.set_xlabel("Date")
ax.set_ylabel("Yield")
ax.set_title("Yields Curve")
ax.legend()
ax.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```
